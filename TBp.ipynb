```python
import pandas as pd
```

```python
df = pd.read_csv("Input_csv/2018hitting_combined.csv")
df
```

```python
#Isolate relevant columns 
df = df[['Player', 'G', 'PA', 'AB', 'H', '2B', '3B', 'HR', 'SB', 'CS', 'BB', 'HBP', 'SH', 'SF']]
df
```

```python
#Create column for singles
df['1B'] = df['H'] - df['2B'] - df['3B'] - df['HR']
df.columns
```

```python
#Reorder columns
df = df[['Player', 'G', 'PA', 'AB', 'H', '1B', '2B', '3B', 'HR', 'SB', 'CS',
       'BB', 'HBP', 'SH', 'SF']]
#https://www.kite.com/python/answers/how-to-reorder-columns-in-a-pandas-dataframe-in-python
df
```

```python
weights = {'1B':1, '2B':2, '3B':3, 'HR':4, 'SB':1, 'CS':-1, 'BB':1, 'HBP':1, 'SH':1, 'SF':1}
df['TBp'] = (weights['1B']*df['1B'] + weights['2B']*df['2B'] + weights['3B']*df['3B'] + weights['HR']*df['HR'] + 
            weights['SB']*df['SB'] + weights['CS']*df['CS'] + weights['BB']*df['BB'] + weights['HBP']*df['HBP'] +
            weights['SH']*df['SH'] + weights['SF']*df['SF'])
df
```

```python
df.sort_values(by = ['TBp'], ascending = False, inplace = True)
df
```

```python
df = df.reset_index(drop = True)
df
```

```python
df.head(20)
```

```python
df['TBp / PA'] = df['TBp']/df['PA']
df.sort_values(by = ['TBp / PA'], ascending = False, inplace = True)
df = df.loc[df['PA'] > 100]
df = df.reset_index(drop = True)
df.head(20)
```

```python
df.to_csv('Output_csv/2018TBp.csv', index = False)
```
